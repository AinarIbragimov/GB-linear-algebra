## Практическое задание. Часть 1

#### 1) Исследовать на линейную зависимость: 
$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)= x−e^{x}$

$x = f_3(x) - f_2(x)$

$f_4(x)=f_3(x)-f_2(x)-f_1(x)$

$f_4(x) - f_3(x) + f_2(x) + f_1(x) = 0$

Векторы линейно независимы

#### 2) Исследовать на линейную зависимость: 
$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)= (x+1)^{2}$

$f_4(x)=x^2 + 2x + 1$

$f_4(x)=f_3(x) + 2f_2(x) + f_1(x) - 1$

$ f_3(x) + 2f_2(x) + f_1(x) - f_4(x) = 1$


Векторы линейно зависимы

$x=(2, 3, 5)=(2,0,0)+(0, 3, 0)+(0, 0, 5)=2\cdot(0, 0, 10)+3\cdot(2, 0, 0)+5\cdot(0, 1, 0)=(6, 5, 20)$


координаты вектора $x$ в стандартном базисе $6$, $5$, $20$.

#### 4) Найти координаты вектора $3x^{2}−2x+2∈R^{3}[x]$:

а) в базисе $1, x, x^{2}$;

б) в базисе $x^{2}, x−1, 1$

Составим вектор из полинома: $[3, -2, 2]$

В базисе $1, x, x^2$: $$(3, 0, 0)\cdot(1, 0, 0) + (0, -2, 0)\cdot(0, x, 0) + (0, 0, 2)\cdot(0, 0, x^2) = (3, -2x, 2x^2)$$

В базисе $x^2, x - 1, 1$: $$(3, 0, 0)\cdot(x^2, 0, 0) + (0, -2, 0)\cdot(0, x - 1, 0) + (0, 0, 2)\cdot(0, 0, 1) = (3x^2, -2x + 2, 2)$$

#### 5) Установить, является ли линейным подпространством:
а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;

б) все векторы, являющиеся линейными комбинациями данных векторов ${ u_{1},u_{2},...,u_{n} }.$

а) возможны три вида векторов: (0, a, b), (a, 0, b) и (0, 0, a).

1)

$(0,a,b)+(0,c,d)=(0,a+c,b+d)$

$\alpha*(0,a,b)=(0,\alpha a,\alpha b)$

2)

$(a,0,b)+(c,0,d)=(a+c,0,b+d)$

$\alpha*(a,0,b)=(\alpha a,0,\alpha b)$

3)

$(0,0,a)+(0,0,c)=(0,0,a+c)$

$\alpha*(0,0,a)=(0,0,\alpha a)$

Во всех трёх случаях полученные векторы также принадлежат указанному в задании множеству всех веторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю, то есть данное множество является подпространством линейного пространства $R^{3}.$

б) $(a,b,c)+(d,e,f)=(a+d,b+e,c+f)$

$\alpha*(a,b,c)=(\alpha a,\alpha b,\alpha c)$

Векторы также принадлежат указанному в задании множеству всех веторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю, то есть данное множество является подпространством линейного пространства $R^{3}.$

#### 4) Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $R^{3}$:

$а) (1,0,0),(0,0,1);$

$б) (1/\sqrt{2},−1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0),(0,0,1);$

$в) (1/2,−1/2,0),(0,1/2,1/2),(0,0,1);$

$г) (1,0,0),(0,1,0),(0,0,1)?$

$а) x=(1,0,0) y=(0,0,1)$

$(x,y) = 1*0+0*0+0*1=0$

$(x,x) = 1*1+0*0+0*0=1$

$(y,y) = 0*0+0*0+1*1=1$

Данные вектора образуют ортонормированный базис в линейном пространстве $R^{3}$

$б) x=(1/\sqrt{2},−1/\sqrt{2},0),y=(1/\sqrt{2},1/\sqrt{2},0),z=(0,0,1)$

$(x,y)=1/\sqrt{2}*1/\sqrt{2}−1/\sqrt{2}*1/\sqrt{2}+0*0=0$

$(x,z)=1/\sqrt{2}*0−1/\sqrt{2}*0+0*1=0$

$(y,z)=1/\sqrt{2}*0+1/\sqrt{2}*0+0*1=0$

$(x,x)=1/\sqrt{2}*1/\sqrt{2}+(−1/\sqrt{2})*(−1/\sqrt{2})+0*0=0.5+0.5+0=1$

$(y,y)=1/\sqrt{2}*1/\sqrt{2}+1/\sqrt{2}*1/\sqrt{2}+0*0=0.5+0.5+0=1$

$(z,z)=0*0+0*0+1*1=1$

Данные вектора образуют ортонормированный базис в линейном пространстве $R^{3}$

$в) x=(1/2,−1/2,0),y=(0,1/2,1/2),z=(0,0,1)$

$(x,y)=\frac{1}{2}*0-\frac{1}{2}*\frac{1}{2}+0*\frac{1}{2}=-\frac{1}{4}\neq0$

Данные вектора не образуют ортонормированный базис в линейном пространстве $R^{3}$

$г) x=(1,0,0),y=(0,1,0),z=(0,0,1)$

$(x,y)=1*0+0*1+0*0=0$

$(x,z)=1*0+0*0+0*1=0$

$(y,z)=0*0+1*0+0*1=0$

$(x,x)=1*1+0*0+0*0=1$

$(y,y)=0*0+1*1+0*0=1$

$(z,z)=0*0+0*0+1*1=1$

Данные вектора образуют ортонормированный базис в линейном пространстве $R^{3}$
